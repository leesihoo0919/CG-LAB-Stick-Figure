```python
import cv2
import numpy as np
import os
import cv2
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
from matplotlib import pyplot as plt

## Distance Transform 적용 방법

# 이미지 로드
image = cv2.imread("./image_ex/sf001782.png", cv2.IMREAD_GRAYSCALE)
print(image)
# 이진화
_, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

# 거리 변환 적용
distance_transform = cv2.distanceTransform(binary_image, cv2.DIST_L2, maskSize=3)

# 거리 변환 결과 출력
# 거리 변환 결과 출력
plt.imshow(distance_transform, cmap='gray')
plt.axis('off')
plt.title('Distance Transform')
plt.show()

```
